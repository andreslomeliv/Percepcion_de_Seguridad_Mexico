## 1. Introducción 

La inseguridad y crimen como una problemática social, es un fenómeno preocupante que responde a diversos procesos y factores particulares de cada región, su estudio desde un enfoque teórico es algo que durante varias décadas atrás se ha abordado bajo distintos paradigmas, y a medida que los métodos de captación y procesamiento de información se han modernizado, también se ha vuelto posible realizar distintos análisis empíricos que buscan dar luz sobre el comportamiento cuantitativo de los indicadores de criminalidad y su relación con el curso de otras variables clave alrededor del mundo. 

En el caso mexicano, un país cuyo porcentaje de personas mayores a 18 años que perciben inseguridad en su entidad durante el primer trimestre de 2018 alcanzó el 78%, un país cuya capital en el 2018 presentó un porcentaje de 40% de mujeres mayores de 18 años víctimas de algún delito, donde solo en el 2017 ocurrieron más de 9 millones de robos en la calle o transporte y donde alrededor del 89% de delitos no es denunciado por sus víctimas (INEGI, ENVIPE), existen muchas conexiones que muestran cómo la criminalidad ha sido un factor asociado a un bajo desarrollo y un lento desempeño económico. La inseguridad en México no es un problema reciente, ya más de una década nos separan de la llamada ‘guerra contra el narcotráfico’ que evidenció la necesidad de atender conscientemente la problemática de la inseguridad en México, aplicando políticas públicas que muestren funcionalidad, y aún prolifera una condición de fallida de los ejes de seguridad en muchas partes del país. Encontramos que existe extensa bibliografía que se plantea examinar empíricamente la problemática de las tasas de criminalidad y seguridad, buscando establecer los determinantes precisos de la inseguridad por medio de diferentes tecnicas de modelacion y selección de variables para cada caso, por ello, proseguimos a hacer una revisión de material que nos pareció relevante para abordar el fenómeno. 

La variable central del análisis de la criminalidad puede ser diferente según el enfoque del problema o la disponibilidad de información, puede usarse el índice delictivo, la tasa de crimen o cualquier variable que mida el grado de proliferación de la inseguridad, pero en cualquier caso, es frecuentemente encontrarnos que la bibliografía establece una fuerte relación de la inseguridad con las condiciones socioeconómicas del entorno donde ocurre el delito, trabajos como el Ralph C. Allen (1996), estudian como las condiciones socioeconómicas tales como la pobreza, el sistema de justicia, la condición familiar y algunas variables del entorno macroeconómico tienen efectos sobre la propensión criminal, en su trabajo, realiza un modelo panel que abarca el periodo de 1959 a 1992, para estudiar cada una de las principales tasas de criminalidad por tipo de crimen en Estados Unidos; encuentra que la variable de pobreza absoluta es significativa y esta negativamente relacionada con las tasas de crimen, debido posiblemente a que los incentivos de cometer un crimen cuando hay ingresos muy bajos es mayor; por su parte la variable que mide la pobreza relativa o desigualdad (índice de gini) muestra la misma relación con la tasa de crimen pero esta variable no es significativa, explica el autor, debido a que los grupos de ingresos altos se preocupan más por su protección, eliminando las oportunidades de crimen entre grupos; por su parte, la variable tasa de desempleo es positivamente significativa para la tasa de asaltos y hurtos, reflejando la misma conexión que en la pobreza; mientras que la variable inflación, que refleja el entorno macroeconómico, muestra una relación positiva con significancia estadística para varios tipos de crímenes como robo hurto y robo de autos, indicando que la inflacion es un motivante del crimen a la propiedad; siguiendo con las variables que describen la situacion sociofamiliar como el porcentaje de poblacion joven de entre 15 y 34 años, el indice de actividad por hogar y el porcentaje de nacimientos de madres solteras, muestran insignificancia estadistica con la tasa de criminalidad; por ultimo, la variable referente al sistema de justicia, que mide la posibilidad de ser aprendido, es el porcentaje de delitos resueltos por cada 100 mil habitantes, y muestra una relacion negativa con significancia solo para hurto y robo de autos, indicando que el sistema de justicia solo ayuda a disminuir crimenes a la propiedad. Estas conclusiones dotan de un carácter más importante a las variables del entorno económico como determinantes del crimen, sobre las condiciones socio familiares. 

Otros trabajos como el de Steven Raphael and Rudolf Winter‐Ebmer (2001) centran su atención en el efecto que tienen las condiciones laborales y  sociales sobre siete tipos de crimen, violentos y no violentos, realiza un modelo panel para los años de 1971 a 1997 para 50 estados de Estados Unidos, su objetivo es establecer el efecto que tienen las oportunidades de empleo, las oportunidades de cometer crímenes, el consumo de alcohol y la respuesta del sistema de justicia sobre el logaritmo del número de crímenes por cada 100 mil habitantes, sus descubrimientos revelan que; la tasa de desempleo tiene un efecto positivo y significativo para todos los tipos de crímenes, pero muestra un efecto mayor sobre los crimenes a la propiedad, mostrando que el desempleo es un incentivo a los crimenes no violentos; a su vez la variable que mide la tasa de encarcelamiento, cifra que refleja la posibilidad de ser atrapado, tiene solo un efecto negativo significativo sólo sobre los crímenes violentos; el consumo de alcohol per cápita es positivo y significativo sólo para los crímenes a la propiedad; se destaca que las tasas de crímenes son mayores en estados donde las áreas metropolitanas son más grandes, también, existe un mayor número de crímenes violentos en los estados donde la proporción de adolescentes y adultos jóvenes es mayor; por su parte, el ingreso por trabajador tiene una relacion negativa y signficativa para todos los tipos de crimenes, reafirmando el efecto de la capacidad economica en la tasa de crimen. En una ampliación del modelo, realizando una regresión en dos etapas, el autor encuentra que la relación entre el desempleo y la tasa de crímenes violentos como el asesinato y la violacion se vuelve no significativa y negativa, concluyendo en que el desempleo y el entorno economico esta mas asociado a crimenes a la propiedad, los cuales no son reducidos por un efecto en el aumento de las tasas de aprensión.

Por último, un estudio aplicado al caso latinoamericano se presenta de mano de los autores Pablo Fajnzylber, Daniel Lederman & Norman Loayza (1998) quienes buscan encontrar los principales determinantes y efectos de las tasas de crimen en latinoamérica, aplicando un modelo tipo panel en con un método generalizado de momentos el periodo 1970-1994 para una muestra de diversos países de la región, haciendo uso de la información del United Nations World Crime Surveys, usan diferentes variables explicativas a nivel pais, que abarcan ámbitos socio demográficos, macroeconómicos, delictivos e institucionales de justicia para ver el comportamiento de la tasa de homicidio intencional y la tasa de robo; los autores encuentran resultados distintos, pero con niveles que especificación de la regresión entre 1 y 3, encontramos que las variables que describen el entorno macroeconómico como la tasa de PIB per cápita y el índice de gini son significativas para la tasa de homicidios y robos, indicando que la desigualdad y la pobreza son incentivos para ambos tipos de crimen; respecto a las variables que describen condiciones sociales como el promedio de los años de escolaridad y la tasa de urbanización resultan no significativas para la tasa de homicidios pero significativos para la tasa de robos, indicando la conexion entre el nivel de desarrollo individual y su conexcion con los crimenes de propiedad ; por su parte, las variables dummy’s delictivas que cuantifican si el pais es un centro podructor y conusmidor de drogas resulta significativa y positiva para la tasa de homicidios y robos, mostrando que la proliferación de un mercado de drogas esta asociado con la ocurrencia de otros crimenes; por su parte las variable de la tasa de criminalidad previa, de homicidio o robo respectivamente, resulta positiva y significativa para ambos modelos, es decir, la tasa de crimenes previos esta asociada a la tasa actual; con respecto a las variables de seguridad como el numeros de elementos de policía por cada 100 mil habitantes y el número de crímenes con sentencia, que solo fue introducida en el modelo de la tasa de robos, encontramos que tienen un pequeño efecto negativo con significancia estadística, es decir, el mejorar la estructura de justicia tiene un ligero efecto reducir la tasa delictiva por propiedad. Sus resultados muestran que hay diversos factores que tienen efectos distintos sobre las tasas delictivas de homicidio y robo, al igual que en el estudio anterior, vemos efectos diferenciados por tipo de crimen, pero es indudable una asociación de las condiciones socioeconómicas de las personas con los crímenes a la propiedad, es decir, la tasa de delictiva de crímenes a la propiedad está asociada con las condiciones socioeconómicas de la población.

Considerando estas aportaciones, vemos la relevancia de definir primero un modelo que considere observaciones a nivel individuo y entidad. A su vez, conscientes de que muchos de los sucesos delictivos se reportan con atraso y muchos otros quedan sin ningún tipo de antecedente, optamos por usar la percepción de seguridad a nivel localidad por persona como variable regresada, esto con el objetivo de capturar con mayor precisión el estado de seguridad por localidad, por vía de la percepción de sus habitantes. 

Optamos, a diferencia de los estudios previos, por incorporar el sexo como una variable importante en la percepcion de seguridad, esto con el objetivo de analizar como la condicion estructural de genero tiene un efecto importante en la condicion de seguridad, tambien añadimos la edad, pensando en como el efecto de la pertenencia a un grupo de edad puede afectar la percepcion de seguridad, y si es significativo para el caso mexicano. Tomamos también el grado de escolaridad individual, que busca definir el posible estrato social al que pertenece la persona, considerando que entre menor sea el grado de educación de las personas menor será el ingreso y la zona percibirá menos seguridad, se considera además la condición laboral de la persona, categorizando si es que trabaja o no, buscando establecer el efecto del ingreso de la localidad sobre la seguridad. En busca de capturar el efecto preciso del ingreso sobre la percepción de seguridad, incorporamos las variables de tipo de vivienda, que por su distribución de información quedaron reducidas a si la persona habita o no en una casa independiente, además, se añadió la variable de si su casa tiene barrera física, indicando cual es el efecto que tiene el tipo de vivienda sobre la percepción de seguridad. Con el fin de incorporar variables que se aproximen a cuál es el grado de inversión infraestructural en la localidad, tenemos las variables de percepción de problemáticas de alumbrado, agua, baches y fugas, con estas variables queremos capturar el efecto de la actividad gubernamental a nivel local y cómo influye en la percepción de seguridad. Las siguientes variables seleccionadas son construidas a partir de una lista de preocupaciones de las personas, estas reflejan si los individuos están preocupados por algún problema de carácter económico y o de seguridad, con esto queremos cuantificar si en el entorno existen mayoritariamente problemáticas de carácter económico o de seguridad. 

Como se puede observar, hasta ahora nuestras variables son de carácter individual en pro de establecer efectos más precisos y condiciones particulares de la percepción de seguridad, pero en búsqueda de establecer regresores que reflejan el efecto que tiene entorno económico sobre la seguridad, incorporamos diversas variables a nivel estatal, que describen el comportamiento del empleo con la tasa desempleo, la actividad económica con el ITAEE, el gasto en seguridad con el porcentaje de gasto estatal destinado a seguridad, la desigualdad con el índice de gini y la pobreza con el porcentaje de personas en situación de pobreza a nivel estatal.

Finalmente, para analizar el impacto de estas variables sobre la percepción de seguridad se estimará un modelo logístico que determine para un individuo si se siente seguro o inseguro en su colonia dadas sus condiciones particulares o las condiciones de la entidad. Para este modelo se propone una estructura de efectos mixtos anidados a nivel entidad que nos permita observar las diferencias en la percepción de seguridad para los diferentes estados de la república. 

## 2. Descripción de los datos utilizados

La principal fuente de información para obtener los datos del modelo proviene de la [Encuesta Nacional de Victimización y Percepción sobre Seguridad Pública](https://www.inegi.org.mx/programas/envipe/2021/#:~:text=La%20Encuesta%20Nacional%20de%20Victimización,Instituto%20Nacional%20de%20Estadística%20y) (ENVIPE 2016-2020), realizada anualmente por Instituto Nacional de Estadística, Geografía e Informática (INEGI). Dicha encuesta es un proyecto estadístico qué busca obtener información sobre las características de la delincuencia en el país, la percepción sobre seguridad pública, así como la realidad de las víctimas de hechos delictivos. El diseño de la muestra es probabilístico; es decir, los resultados de la encuesta se generalizan a toda la población, teniendo como última unidad de selección la vivienda y como unidades de observación a los hogares y a las personas de 18 años cumplidos o más.
 
La ENVIPE genera información con representatividad a nivel nacional y estatal que permitirá analizar la prevalencia delictiva, los niveles de incidencia delictiva, la cifra negra, percepción de la seguridad pública, así como el desempeño de las instituciones a cargo de la seguridad pública y la justicia. 

Además de esta base también se usaron datos a nivel estatal sobre desocupación, actividad económica, egresos en seguridad pública, desigualdad y pobreza. Con las fuentes de información previamente especificadas, procedemos a describir brevemente las variables utilizadas en la creación de nuestro modelo. Con la finalidad de hacer más intuitiva y breve esta sección, omitiremos las variables renglon, hogar,	 viv_sel y upm, que son variables llave o identificadores que fueron usados para la creación de nuestra base final, además, separamos la descripción de las variables en dos conjuntos; uno correspondiente a los datos de las observaciones individuales de nuestra muestra, y otro conjunto referente a las variables que representan información a nivel de entidad federativa, también hay que decir que al ser un modelo de datos panel existe un identificador temporal por años, señalado como ```año``` que va del 2016 al 2020 para todas las observaciones . Para todas las variables se especifica su fuente de origen, tipo de dato y su relación en la forma funcional del modelo.

### 2.2 Variables individuales

Empezando con las variables concernientes a las características de las unidades encuestadas provenientes de la ENVIPE, tenemos primero la variable que enmarca nuestro objetivo de pronóstico, es decir, la variable respuesta, definida como ```perc_seguridad``` que expresa individualmente la percepción de seguridad a un nivel de colonia, con una forma dicotómica arroja un valor de 1 si la persona percibe seguridad en su colonia y 0 si no es así. Siguiendo con las variables independientes o regresoras, tenemos las variables demograficas definidas como; ```sexo``` con una division dicotmica que marca un valor 1 si el individuo es masculino y 0 si es femenino; y ```edad```, con forma discreta, que señala la edad en años de los individuos en la muestra con mayoría de edad.

Siguiendo con las variables explicativas, se describen las variables correspondientes a las condiciones educativas y laborales de los individuos en la muestra. Primero tenemos la variable ordinal ```escolaridad```, que señala categóricamente el nivel de estudio que tiene el individuo (0: ninguno, 1: precolar, 2: primaria, 3: secundaria, 4: carrera técnica con secundaria, 5: normal básica, 6: preparatoria, 7: carrera técnica con preparatoria, 8: licenciatura, 9: maestría o doctorado). Luego tenemos las variables dicotomicas ```con_trabajo```, ```no_puede_trabajar``` y ```no_trabajo``` que señalan el estado de ocupación de los encuestados. Estas tres variables son construidas bajo tres criterios de agrupación; si el individuo en la semana previa a la encuesta trabajó o tuvo trabajo aunque no haya trabajado, se envía a la primer variable, que indica 1 si cumple cualquiera de estas condiciones o 0 si no es así; por otro lado, si el individuo es estudiante, se dedica a quehaceres del hogar, esta incapacitado o jubilado sera considerando en la segunda variable, que marca 1 si cumple con alguna de estas condiciones y 0 de ser lo contraio; por último, si el individuo no trabajo o busco trabajo en la semana previa a la escuesta se considera en la tercer variable, que de igual forma, arroja 1 en caso de cumplir con estas condiciones y 0 en caso de que no. 

Las siguientes variables describen la situación particular de vivienda de las observaciones, estas variables dicotomicas son ```casa_independiente```, ```departamento```, ```vecindad```, ```cuarto_azotea``` y ```local_no_habitacional``` y de forma excluyente señalan un valor 1 cuando el individuo cumple con la condición de vivir en cada tipo de vivienda descrita, respectivamente y con 0 de lo contrario, además, existe la variable ```barrera_vivienda``` que describe de forma dicotómica si la vivienda del individuo tiene algún tipo de barrera física que impida llegar de forma directa a la puerta.  

Siguiendo con la descripción, tenemos un subconjunto de variables individuales no excluyentes que señalan la percepción de problemática infraestructural en su localidad, estas son ```falta_alumbrado```, ```falta_agua```, ```baches_fugas```, que de manera dicotómica, indican con 1 la percepción afirmativa del problema y con 0 la negativa. 

Por último, ```bienestar_econ``` y ```seguridad``` describen el temor que los individuos tienen hacia dos rubros particulares, el económico y el de seguridad respectivamente. Son variables dicotómicas no excluyentes construidas a partir de otros datos; la primer variable es un agrupado de que el individuo está preocupado por sus condiciones económicas y materiales, esta variable tendrá un valor 1 cuando al menos uno de los siguientes fenómenos le preocupe: pobreza, desempleo, inflación y escasez de agua, y 0 de lo contrario; por su parte, la segunda variable marcará un valor 1 cuando al menos uno de los siguientes fenómenos le preocupe: narcotráfico, inseguridad, corrupción e impunidad, y 0 de lo contrario. 

### 2.3 Variables estatales

El resto de las variables independientes tienen que ver con agregadoas a nivel Entidad Federativa obtenidos de diferentes instituciones. La variable ```desocupación``` proviene de la Encuesta Nacional de Ocupación y Empleo (ENOE) y se refiere a la tasa de desocupación de la Entidad Federativa. Esta variable se obtuvo dividiendo la población desocupada entre la población económicamente activa. Ambas series son proporcionadas por el INEGI y se encuentran en el [Banco de Indicadores](https://www.inegi.org.mx/app/indicadores/). 
Por otro lado, la variable ```itaee``` se refiere al Indicador Trimestral de Actividad Económica Estatal. Esta serie es calculada por el INEGI y se obtiene en el [Banco de información Económica](https://www.inegi.gob.mx/app/indicadores/?tm=0&t=10000215#D10000215). Tanto la tasa de desocupación como el ITAEE son series trimestrales por lo que se obtuvo el promedio simple de los cuatro trimestres para anualizar las variables. 

La variable ```pct_seguridad``` se refiere al porcentaje de los egresos estatales que se destinaron a la seguriad pública. Esta se obtuvo sumando todas las partidas genéricas que tuvieran en su descripción las palabras *seguridad pública* y dividiendo entre el total de los egresos para cada año. Los datos con los cuales se calculó esta variable son proporcionados por el INEGI en los tabulados de [Finanzas públicas estatales y municipales](https://www.inegi.org.mx/programas/finanzas/).  

Finalmente, las últimas dos variables provienen de la Encuesta Nacional de Ingresos y Gastos de los Hogares (ENIGH) y son calculadas por el [Consejo Nacional de Evaluación de la Política de Desarrollo Social](https://www.coneval.org.mx/Medicion/MP/Paginas/Pobreza_2020.aspx) (CONEVAL). La variable ```gini``` se refiere al Coeficiente de Gini de la entidad y mide la desigualdad en los ingresos. La variable ```pobreza``` se refiere al cálculo que realiza el CONEVAL de la pobreza multidimencional y representa el porcentaje de la población que se encuentra en situación de pobreza en la entidad. Ya que la ENIGH se publica cada dos años, dentro del periodo de estudio solo contamos con valores para el 2016, 2018 y 2020. Los valores para el 2017 y el 2019 se obtuvieron calculando el promedio simple de los dos años entre los que se encuentran. Es decir: el valor para el año 2017 es el promedio del valor en el 2016 y el 2018 mientras que el valor para el 2019 es el promedio del 2018 y el 2020.

## 4. Estimación del modelo

En esta sección estimaremos un modelo de elección binaria que se ajuste de la mejor manera a nuestros datos. Para esto se consideró como función de densidad una función logística y para los regresores se eligió una estructura de efectos mixtos considerando un sistema de ecuaciones de dos niveles: una función a nivel individual y otras anidadas por entidad Federativa. De esta manera, la forma funcional para la variable latente será la siguiente: 

$$ Y* = log(\frac{P}{1-P}) = X\beta + Z\theta + \epsilon $$

donde la matriz $X$ incluye los regresores con efectos fijos y la matriz $Z$ incluye los regresores con efectos aleatorios. De manera más concreta podemos escribir nuestro modelo como un sistema de ecuaciones con dos niveles de anidamiento:
 
$$ N_1: Y_{ij}* = \beta_{0j} + \sum_{k=1}^{K-1}\beta_k X_{ki} + \beta_{Kj} Sexo_i + \sum_{m=1}^{M}\theta_m Z_{mj} + {\epsilon}_i $$
$$ N_2: \beta_{0j} = \theta_{00} + r_{0j} $$
$$ N_2: \beta_{kj} = \theta_{k0} + r_{kj} $$

donde $\beta_{0j}$, $j \in {1,...,32}$ es un intercepto aleatorio anidado a nivel entidad, las variables $X_k$, $k \in {1,2,..,K-1}$ son los regresores a nivel individual y con efectos fijos mientras que $Sexo_i$ es una variable a nivel individual pero con pendiente aleatoria $\beta_{Kj}$. Por otro lado, las variables $Z_{mj}$, $m \in {1,...,M}$ son variables a nivel entidad con efectos fijos. Resolviendo el modelo obtenemos la siguiente ecuación para la variable latente:

$$ Y_{ij}* = \theta_{00} + \sum_{k=1}^{K-1}\beta_k X_{ki} + \theta_{k0} Sexo_i + \sum_{m=1}^{M}\theta_{0m}Z_{mj} + r_{0j} + r_{kj} Sexo_i + {\epsilon}_i $$

Finalmente, ya que nuestra variable dependiente es binaria, la función a estimar tiene la siguiente forma:

$$ Y_{i} = \frac{e^{Y*}}{1 - e^{Y*}} $$

Esta estructura de anidamiento se contrastará con un modelo logístico simple sin efectos mixtos en el que la matriz $X$ incluye también las variables a nivel entidad. De esta forma podremos observar mejor si la estructura de anidamiento añade información omitida al modelo. Para este modelo base la forma funcional a estimar será la siguiente:

$$ Y_{i} = \frac{e^{Xi\beta + {\epsilon}_i}}{1 - e^{Xi\beta + {\epsilon}_i}} $$

donde 
$$ Y_i = \begin{cases} 1 & \text{Seguro}\\   0 & \text{Inseguro} \end{cases} $$

````python
#! pip install pymer4
# por alguna razón la primera vez marca error pero la segunda no
from pymer4.models import Lmer
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix
import scipy
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
import warnings
from INEGIpy import MarcoGeoestadistico
warnings.filterwarnings("ignore")
````

````python
# modelo nulo para calcular el ICC
modelo_nulo = Lmer('perc_seguridad ~ 1 + (1|cve_agee)', data=df, family='binomial')
modelo_nulo.fit()
````

````python
df.itaee = StandardScaler().fit_transform(df.itaee.values.reshape(-1, 1))
df.pobreza = df.pobreza/100
````

````python
df = df.sample(frac=1, random_state=111).reset_index(drop=True) # revolvemos el DF para separarlo en uno de entrenamiento y otro de prueba
idx = round(df.shape[0]*0.20)
df_test = df.iloc[:idx,:]
df_train = df.iloc[idx:,:]
````

````python
y_formula = 'perc_seguridad ~ '
x_formula = 'bienestar_econ + seguridad + falta_alumbrado + falta_agua + baches_fugas + (sexo-1|cve_agee) + edad + escolaridad + '
x_formula += 'barrera_vivienda + casa_independiente + con_trabajo + desocupacion + itaee + pct_seguridad + gini + pobreza + '
x_formula += '(1|cve_agee)'
formula = y_formula + x_formula
````

````python
model = Lmer(formula, data=df_train, family='binomial')
model.fit()
````

````python
preds = model.predict(df_test, verify_predictions=False)
df_test['preds'] = preds
df_test['preds'] = df_test.preds.apply(lambda x: 0 if x < 0.5 else 1) 
````

````python
def print_metrics(y_obs, y_pred):
    tn, fp, fn, tp = confusion_matrix(y_obs, y_pred).ravel()
    print('Exactitud: {:.2%}'.format((tp + tn) / (tp + tn + fp + fn)))
    print('Precisión: {:.2%}'.format((tp) / (tp + fp)))
    print('Sensibilidad: {:.2%}'.format((tp) / (tp + fn)))
    
    
print_metrics(df_test.perc_seguridad, df_test.preds)
````

````python
ranef = model.ranef
ranef.transpose()
````

````python
ranef.min()
````

Del modelo anidado podemos ver que se adapta bien a los datos y logra clasificar correctamente la mayor parte de las nuevas observaciones. Igual destaca que todas las variables excepto el Coeficiente de Gini sean estadísticamente significativas. No obstante, los problemas de convergencia del modelo llaman la atención. Esto puede a que alguno o todos los parámetros restringidos de los efectos aleatorios son cero o muy cercanos a cero, aunque no parece ser el caso para los parámetros estimados. De igual forma el problema se puede deber a la escala de las variables o a un problema con el optimizador. Dadas las limitaciones del la paquetería para probar otros optimizadores y a que los problemas de convergencia no parecen afectar la capacidad de predicción del modelo, continuaremos usando la estimación resultante. 

Para comparar mejor nuestro modelo estimaremos un modelo logístico sin efecttos mixtos. Es decir, se estimará una regresión pool para observar cuánta información es agregada por la estrectura de los términos de error.

````python
x_cols = ['bienestar_econ','seguridad','falta_alumbrado','falta_agua','baches_fugas','sexo','edad','escolaridad',
          'barrera_vivienda','casa_independiente','con_trabajo','desocupacion','itaee','pct_seguridad','gini','pobreza']
pool_model = sm.Logit(df_train['perc_seguridad'], df_train[x_cols])
pool_model = pool_model.fit()
````

````python
pool_model.summary2()
````

````python
pool_model.save("logit_pool.pickle")
````

````python
df_test['pool_preds'] = pool_model.predict(df_test[x_cols])
````

````python
df_test.pool_preds = df_test.pool_preds.apply(lambda x: 0 if x < 0.5 else 1)
````

````python
print_metrics(df_test.perc_seguridad, df_test.pool_preds)
````

Podemos ver que la regresión pool muestra métricas similares al modelo anidado. Entre las principales diferencias están los niveles de significancia de los coeficientes, ya que la variable ```barrera_vivienda``` aumenta su $p-valor$ aunque se mantiene significativa con un 90% de confianza mientras que la variable ```gini``` se vuelve significativa y la variable ```pct_seguridad``` pierde significancia. También podemos observar que el AIC y las medidas de exactitud, precisión y sensibilidad muestran que la estructura de los términos de error sí logra agregar más  información al modelo aunque de manera marginal. 

Por su mayor capacidad de predicción y por fines de la investigación se decidió utilizar el modelo de efectos mixtos para el análisis de los resultados. Tal como se esperaba, a mayor desigualdad, pobreza y desocupación para la entidad menor será la probabilidad de que los habitantes se sientan seguros en su colonia.
